```python
# OUTLIER DETECTION FOR MULTIPLE COLUMNS USING THE IQR METHOD
#numeric features: ['team', 'targeted_productivity', 'smv', 'wip', 'over_time', 'incentive', 'idle_time', 'idle_men', 'no_of_style_change', 'no_of_workers', 'actual_productivity']
# Select all the numerical columns 
numerical_cols = ['team', 'targeted_productivity', 'smv', 'wip', 'over_time', 'incentive', 'idle_time', 'idle_men', 'no_of_style_change', 'no_of_workers', 'actual_productivity']

print("Detecting outliers in these columns:")
print(numerical_cols)
print("-" * 100)

# Dictionary to store outlier indices per column
outliers_dict = {}
outlier_counts = []

# Apply IQR rule to each numerical column
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outlier rows
    col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
    
    outliers_dict[col] = col_outliers
    outlier_counts.append(len(col_outliers))
    
    print(f"{col:20}  {len(col_outliers):3} outliers "
          f"(bounds: {lower_bound:.2f} to {upper_bound:.2f})")

# Total unique rows with at least ONE outlier
all_outlier_rows = pd.Index([])
for idx in outliers_dict.values():
    all_outlier_rows = all_outlier_rows.union(idx)

print("-" * 100)
print(f"Total unique rows with outliers: {len(all_outlier_rows)} out of {len(df)}")
print(f"Percentage: {len(all_outlier_rows)/len(df)*100:.2f}%")

```

```python
# Cap extreme values at 99th percentile (keeps all 1197 rows!)
df_clean = df.copy()

for col in ['targeted_productivity', 'wip', 'over_time', 'incentive', 'idle_time', 'idle_men', 'no_of_style_change', 'actual_productivity']:
    upper = df_clean[col].quantile(0.99)
    df_clean[col] = df_clean[col].clip(upper=upper)

print("Outliers capped at 99th percentile")
print("Shape before:", df.shape, " after:", df_clean.shape)
```

## Data splitting into training and testing

```python
print(numeric_df.head())
```

```python
# Features and target
train = numeric_df.drop('actual_productivity', axis=1)
target = numeric_df['actual_productivity']
```

```python
# Split into train/test (80/20)
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.2, random_state=42)
```

```python
# model selection
# import and train base individual base learners (e.g., Random Forest Regressor, logistic regression, SVR).
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "KNN Regressor": KNeighborsRegressor(n_neighbors=5)
}
```

```python
# training and evaluation for each model

results = {}

for name, model in models.items():
    # Train
    model.fit(X_train, y_train)
    
    # Predict on TEST set (or change to X_val if you have validation)
    y_pred = model.predict(X_test)
    
    # UNIVERSAL RMSE  works on ALL sklearn versions
    try:
        rmse = mean_squared_error(y_test, y_pred, squared=False)
    except TypeError:
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    # R score
    r2 = r2_score(y_test, y_pred)
    
    # Store results
    results[name] = {'RMSE': rmse, 'R': r2}
    
    # print results
    print(f"{name:15}  RMSE: {rmse:.4f} | R: {r2:.4f}")

# FINAL RESULTS TABLE
results_df = pd.DataFrame(results).T
results_df = results_df.sort_values('RMSE')
print("-" * 50)
print("FINAL RANKING (Best to Worst):")
print(results_df.round(6))
print("-" * 50)

    
```

## Using ensemble methods

### Using Bagging (Bootstrap Aggregation)

```python
# Bagging Regressor using Decision Tree as the base estimator
bagging_tree = BaggingRegressor(
    estimator=DecisionTreeRegressor(random_state=42),
    n_estimators=100,
    bootstrap=True,
    n_jobs=-1,              # Use all CPU cores for faster training
    random_state=42
)

# Train the model
bagging_tree.fit(X_train, y_train)

# Make predictions
y_pred_bag = bagging_tree.predict(X_test)

# Evaluate performance
rmse = np.sqrt(mean_squared_error(y_test, y_pred_bag))
r2 = r2_score(y_test, y_pred_bag)

# Display results
print("Bagging Decision Tree Results")
print(f"RMSE: {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```

```python
# AdaBoostRegressor using Decision tree as the base estimator
Adaboost_tree = AdaBoostRegressor(
    estimator=DecisionTreeRegressor(random_state=42),
    n_estimators=100,
    learning_rate=0.1,
    random_state=42                
)
# Train the model
Adaboost_tree.fit(X_train, y_train)

# Make predictions
y_pred_boost = Adaboost_tree.predict(X_test)

# Evaluate performance
rmse = np.sqrt(mean_squared_error(y_test, y_pred_boost))
r2 = r2_score(y_test, y_pred_boost)

# Display results
print(" AdaBoostRegressor Results")
print(f"RMSE: {rmse:.4f}")
print(f"R Score: {r2:.4f}")
```

## Discussion: Strengths & Weaknesses of the Ensemble Methods Used

**Bagging Decision Tree**  
**Strengths:**  
- Lowest RMSE (0.1088)  10% more accurate than AdaBoost  
- Highest R (0.554)  captures 55.4% of target variance  
- Reduces variance through bootstrap averaging  extremely robust to noise  
- Parallel training  fast and scalable  

**Weaknesses:**  
- Slightly higher memory usage (multiple full trees)  
- Less interpretable than a single tree  

**AdaBoostRegressor**  
**Strengths:**  
- Can focus on hard examples sequentially  

**Weaknesses (dominant here):**  
- RMSE 0.1206  11% worse than Bagging  
- R only 0.452  explains 18% less variance  
- Highly sensitive to outliers/noise  overfits by chasing noisy points  
- Sequential  slow and fragile in real-world data  

**Conclusion:**  
Bagging decisively outperforms AdaBoost on this dataset. Its strength lies in variance reduction, while AdaBoosts sequential error-chasing became its downfall in the presence of noise.
